## <font color = blue>I. WHAT IS OCR ?</font>

* Optical Character Recognition (OCR) is the process that converts an image of text into a machine-readable text format.
* For example, if you scan a form or a receipt, your computer saves the scan as an image file. You cannot use a text editor to edit, search, or count the words in the image file. However, you can use OCR to convert the image into a text document with its contents stored as text data.<br>
<br>
* In our study, we used doctr library.
* While using doctr extractor, no image preprocessing is necessary. This is done by the doctr extractor itself.<br>
https://mindee.com/product/doctr

## <font color = blue>II. TEXT EXTRACTION</font>
> A. LIBRARIES

```python
import pandas as pd

from helpers import image_to_string

from time import time

import tensorflow as tf
tf.get_logger().setLevel('ERROR')
```

> B. OCR EXTRACTION

```python
df = pd.read_csv('../data/data_with_meta.csv')
df.head()
```

**Add text field to the original data frame**

```python
# Initialization of a timer
t0 = time()

df['text'] = df['filename'].apply(lambda x: image_to_string(filename = x))

# Calulation of training time
t1 = time() - t0
print("OCR extraction for all dataset in {} minutes".format(t1 // 60))
```
